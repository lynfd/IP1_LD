# 4

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales.head()
```

# 5

```python
car_sales['Price']=car_sales['Price'].str.replace('$','').str.replace(',','').astype('int64')
car_sales
```

# 6

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

# 7

```python
car_sales["Sale Date"] = pd.date_range("26/9/2024", periods=len(car_sales))
car_sales.head()
```

# 8

```python
sns.lineplot(x=car_sales['Total Sales'],y=car_sales['Price'])
plt.title("Price vs Total Sales Line Plot")
```

# 9

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
plt.title("Odometer vs Price Scatter Plot")
```

# 10

```python
sns.barplot(x=car_sales['Make'],y=car_sales['Odometer (KM)'])
plt.title("Make vs Odometer Bar Plot")
```

# 11

```python
sns.histplot(car_sales['Make'])
```

# 12

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sns.lineplot(y='Price', x='Total Sales', data=car_sales, ax=axes[0, 0])

axes[0, 0].set_title('Price vs Total Sales Line Plot')


sns.scatterplot(x='Odometer (KM)',y='Price', data=car_sales, ax=axes[0, 1])

axes[0, 1].set_title('Odometer vs Price Scatter Plot')


sns.barplot(x=car_sales['Make'],y=car_sales['Odometer (KM)'], data=car_sales, ax=axes[1, 0])

axes[1, 0].set_title('Make vs Odometer Bar Plot')

sns.histplot(x=car_sales['Make'], data=car_sales, ax=axes[1, 1])
axes[1, 1].set_title('Histogram of Car Make')
```

# 13

```python
car_sales.boxplot(by="Colour",column="Total Sales", grid=False)
ylabel='Total Sales'
plt.show()
```

# 14

```python
sns.lmplot(y='Odometer (KM)', x='Total Sales', data=car_sales, hue='Make', ci=None, markers=['o','s','D','^'],
           palette='Set1')
plt.title("Linear Regression of Total Sales vs Odometer by Car Make")
```

# 15

```python
sns.pairplot(car_sales[['Total Sales', 'Price', 'Odometer (KM)']])
```

# 16

```python
car_sales.drop('Doors', inplace=True, axis=1) 
corr_matrix=car_sales.select_dtypes('number').corr()
corr_matrix.head()
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation between Total Sales")
```
